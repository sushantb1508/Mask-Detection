```python
from keras.models import load_model
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
```

```python
model=load_model('mask_detector.model')
img=image.load_img('C:/Users/KIIT/Desktop/Face-Mask-Detection-master/mask dataset/Test/without_mask/0_0_anhu_0211.jpg',target_size=(224,224))
x=image.img_to_array(img)
x.shape
x=x/255.0
reshaped=np.reshape(x,(1,224,224,3))
result=model.predict(reshaped)
label=np.argmax(result,axis=1)[0]
label
```

```python
if label == 0:
    print('With Mask')
else:
    print('Without Mask')
```

```python
import datetime
mymodel=load_model('mask_detector.model')

cap=cv2.VideoCapture(0)
face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

while cap.isOpened():
    _,img=cap.read()
    face=face_cascade.detectMultiScale(img,scaleFactor=1.1,minNeighbors=1)
    for(x,y,w,h) in face:
        face_img = img[y:y+h, x:x+w]
        cv2.imwrite('temp.jpg',face_img)
        test_image=image.load_img('temp.jpg',target_size=(224,224,3))
        test_image=image.img_to_array(test_image)
        test_image=test_image/255
        reshaped=np.reshape(test_image,(1,224,224,3))
        result=model.predict(reshaped)
        label=np.argmax(result,axis=1)[0]
        print(label)
        if label==1:
            cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),3)
            cv2.putText(img,'NO MASK',((x+w)//2,y+h+20),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),3)
        else:
            cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)
            cv2.putText(img,'MASK',((x+w)//2,y+h+20),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),3)
        datet=str(datetime.datetime.now())
        cv2.putText(img,datet,(400,450),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,255,255),1)
          
    cv2.imshow('img',img)
    
    if cv2.waitKey(1)==ord('q'):
        break
    
cap.release()
cv2.destroyAllWindows()
```
